A solution was found [here](https://fabianlee.org/2019/01/13/ubuntu-apt-get-error-yarn-signature-verification/) and relates to updating your expired key.

```
# add latest key
curl -sS https://dl.yarnpkg.com/debian/pubkey.gpg | sudo apt-key add -

# try refresh again
sudo apt-get update
```
