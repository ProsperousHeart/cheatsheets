# How To SSH From Windows 10 to VirtualBox Linux VM

STEP 1:  Ensure you have the following adapters enabled:

1. Bridged Adapter


2. Host-Only Adapter (for VirtualBox)


Once these are enabled, you can test your connection by completing the following ...

## On the Linux VM

Turning on the VM and run:  `ifconfig`

You will then see something like this:



The **enp0s8** will likely be your new network option. Both it and **enp0s3** must be running with an IP address.

Make note of the IP for **enp0s8** - this will be what you SSH to from your local Windows machine the VM is running on. In my instance, it is:  `192.168.56.101`

Run `route -n` - should look similar to:



The **G** indicates your default gateway. If you do not have it, you will need to troubleshoot. (Cheatsheet to come.)

You'll notice the default gateway is on the same network as the newer internet connection.

## On Your Local Windows

For your local Windows (which the VM is running on) try to ping the **enp0s8** IP address.

### Using WSL (Windows Subsystem for Linux)

If this is unfamiliar to you, [check this out](https://prosperousheart.com/windows-terminal/). But know that putty is all you need.

If you want to use WSL, then open your ubuntu terminal to try to [SSH to your VM](https://opensource.com/article/20/9/ssh)'s **enp0s8** IP address.
